# Recommendation (show top rating)

```python
import pandas as pd
```

```python
# Load the anime list
pred_users = pd.read_csv("D:/3. Projects/AWS Sagemaker/practice/S3/sagemaker-tutorial-rnd/Recommendation/pred_users_anime_df.csv").iloc[:,1:]
anime_data = pd.read_csv("D:/3. Projects/AWS Sagemaker/practice/S3/sagemaker-tutorial-rnd/Recommendation/anime_data.csv")[
    ["MAL_ID", "Name", "Genres"]
]
```

```python
# Join pred_user and anime data
pred_users_df = pd.merge(pred_users, anime_data, how="left", left_on="anime_id", right_on="MAL_ID")
# Sort by the rating
pred_users_df = pred_users_df.sort_values("pred_rating", ascending=False).dropna()

# Show the top 10 recommended anime titles based on the predicted ratings for this user
pred_users_df.head(10)
```

```python
# Show top recommended action genre for this user
pred_users_df[pred_users_df["Genres"].str.contains("Action")].head(10)
```
